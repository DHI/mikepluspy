# Run MIKE+ Import Tool

This notebook shows how to run tools by using the import tool as an example.

Refer to the documentation for details on other tools.

```python
import mikeplus as mp
from mikeplus.tools import ImportTool

db = mp.open("../tests/testdata/import/import.sqlite")
db
```

```python
# The configuration file can be saved from MIKE+ Import/Export tool
import_tool = ImportTool("../tests/testdata/import/config.xml", db)
import_tool.run()
```
